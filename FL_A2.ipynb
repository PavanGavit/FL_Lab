```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
# Sample student performance dataset
data = {
    "study_hours": [2, 4, 6, 8, 10, np.nan, 5, 7, 3, 9],
    "attendance": [65, 70, 80, 90, 95, 85, np.nan, 88, 72, 92],
    "internal_marks": [40, 50, 60, 70, 85, 75, 55, np.nan, 45, 80],
    "final_score": [45, 55, 65, 75, 88, 78, 60, 82, 50, 85]
}

df = pd.DataFrame(data)
df
```

```python
# Fill missing values with column mean
df = df.fillna(df.mean())

df
```

```python
X = df.drop("final_score", axis=1)
y = df["final_score"]
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
def split_into_clients(X, y, num_clients=3):
    client_data = {}
    data_per_client = len(X) // num_clients

    for i in range(num_clients):
        start = i * data_per_client
        end = (i + 1) * data_per_client if i != num_clients - 1 else len(X)

        client_data[f"client_{i+1}"] = {
            "X": X.iloc[start:end],
            "y": y.iloc[start:end]
        }

    return client_data
```

```python
num_clients = 3
clients = split_into_clients(X_train, y_train, num_clients)

# Display client data sizes
for client, data in clients.items():
    print(client, "-> samples:", len(data["X"]))
```
